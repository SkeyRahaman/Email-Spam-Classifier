In this project i am going to use a folder containing a lot of emails as a text file and haveing an extentiom of (.ham.txt)
for important emails and having an extention of (.spam.txt) for spam emails.I will use "tqdm" library to get a progression bar for each "for" loops in this project.

```python
import numpy as np
import pandas as pd
import os
from collections import Counter
from tqdm import tqdm
```

Here i am storing the path of every email text files in a variable called "folder".

```python
folder = 'email/'
```

```python
files = []
```

Now we will be using a function called "listdir()" which will return a list containing the names of the entries in the directory given by path.

```python
files = os.listdir(folder)
```

```python
files[0:5]
```

Now i will create another list which will contain the full path for every email text file. This is done by a for loop. Parent path was stored in "folder". Hense here we only need to add the file name to "folder" and then we will get the full path. So here i will use a for loop to do the same.

```python
emails = [folder + file for file in files]
```

```python
emails[0:5]
```

Then i will creat a list named "words" which will contail all the words in every email text file. That means the email text file contain a string of that email so what i will do is, i will split the string by (" ") and add each and every word to the list "words" . And will do the same with all the emails. Hense this will take some time to complete .

```python
words = []
for email in tqdm(emails):
    f = open(email,encoding ='latin-1')
    blob = f.read()
    words += blob.split(" ")
```

```python
len(words)
```

Now comes an important part. All the words that are present in "words" is not alphabetc word like "9po" this is a useless entry right?. So i will be removing them With a string builtin method called "isalpha()" which returns "True" if all character in the given string is alphabet. Basically i am removing all the numbered words.

```python
for i in tqdm(range(len(words))):
    if not words[i].isalpha():
        words[i]=""
```

Here i will use a function called Counter that takes a list and returns a dictionary which contains elements of the list as keys and number of their repetation as values.

```python
words_dict = Counter(words)
```

```python
words_dict
```

```python
len(words_dict)
```

As an empty place is not of my use , I will be removing the empty entry.

```python
del words_dict[""]
```

```python
words_dict = words_dict.most_common(3000)
```

```python
words_dict
```

```python
len(words_dict)
```

```python
features = []
target = []
for email in tqdm(emails):
    f = open(email,encoding = "latin-1")
    blob = f.read().split(" ")
    
    data = []
    for i in words_dict:
        data.append(blob .count(i[0]))
    
    features.append(data)
    if 'spam' in email:
        target.append(0)
    if 'ham' in email:
        target.append(1)
```

```python
X = np.array(features)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state = 7)
```

```python
from  sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
```
